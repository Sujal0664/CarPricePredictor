```python
import pandas as pd

df = pd.read_csv("./Car_Dataset_Processed.csv")

df.columns
```

```python
df['fuel_type'].value_counts()
```

```python
d1 = {'Petrol':0, 'Diesel': 1, 'CNG': 2}
df['fuel_type'] = df['fuel_type'].map(d1)
```

```python
df['fuel_type'].value_counts()
```

```python
df['ownsership'].value_counts()
```

```python
df['insurance_validity'].value_counts()
```

```python
d2 = {'Comprehensive': 0, 'Third Party insurance':1, 'Zero Dep': 2, 'Third Party': 1, 'Not Available':3 }
df['insurance_validity'] = df['insurance_validity'].map(d2)
```

```python
d3 = {'First Owner': 1, 'Second Owner':2, 'Third Owner': 3, 'Fifth Owner': 5, 'Fourth Owner': 4 }
df['ownsership'] = df['ownsership'].map(d3)
```

```python
df['ownsership'].value_counts()
```

```python
df['transmission'].value_counts()
```

```python
d4 = {'Manual': 0, 'Automatic':1 }
df['transmission'] = df['transmission'].map(d4)
```

```python
df['transmission'].value_counts()
```

```python
df.head()
```

```python
df['registration_year']
```

```python
df['registration_year'] = df['registration_year'].apply(lambda x: str(x)[-2:])
print(df['registration_year'].head())
```

```python
X = df[['insurance_validity', 'fuel_type', 'kms_driven', 'ownsership', 'transmission']]
Y = df['price(in lakhs)']

X.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
models = {'LinReg': LinearRegression(), 'KNN-3': KNeighborsRegressor(n_neighbors=3), 'SVM-Poly': SVR(kernel='poly')}

for name,model in models.items():
  model.fit(X,Y)
  sc = model.score(X,Y)
  print(name, sc)

```

```python
final_model = models['KNN-3']
```

```python
insurance_validity = input("enter insurance validity: ")
fuel_type = input("enter fuel type: ")
kms_driven = input("enter kms driven:")
ownsership = input("enter ownsership: ")
transmission = input("enter transmission: ")

insurance_validity = d2[insurance_validity]
fuel_type = d1[fuel_type]
ownsership = d3[ownsership]
transmission = d4[transmission]

kms_driven = int(kms_driven)

test = [[insurance_validity, fuel_type, kms_driven, ownsership, transmission]]
yp  = int(final_model.predict(test)[0])
print("Predicted Price of the car is {} Rs.", format(yp))
```
